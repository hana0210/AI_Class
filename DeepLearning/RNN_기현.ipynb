```python
from keras.datasets import imdb
from sklearn.model_selection import train_test_split
(X_train, y_train), (X_test, y_test)  = imdb.load_data(num_words=200)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=.2, random_state=42)
```

```python
word_to_index = imdb.get_word_index()
```

```python
index_to_word = {v + 3: k for k, v in word_to_index.items()}
```

```python
# 0, 1, 2, 3лІҲм—җ н•ҙлӢ№н•ҳлҠ” мҳҲм•Ҫм–ҙ м§Җм •
index_to_word[0] = "<PAD>"    # нҢЁл”© (кёёмқҙ л§һм¶”кё°мҡ©)
index_to_word[1] = "<START>"  # л¬ёмһҘ мӢңмһ‘ нҶ нҒ°
index_to_word[2] = "<UNK>"    # мӮ¬м „м—җ м—ҶлҠ” лӢЁм–ҙ (num_words м ңн•ңмңјлЎң мқён•ҙ нғҲлқҪн•ң лӢЁм–ҙ)
index_to_word[3] = "<UNUSED>"
```

```python
decoded_review = ' '.join([index_to_word.get(i, '?') for i in X_train[0]])
```

```python
from keras.preprocessing.sequence import pad_sequences
X_train = pad_sequences(X_train, maxlen=100)
X_val = pad_sequences(X_val, maxlen=100)
X_test = pad_sequences(X_test, maxlen=100)
```

```python
X_train[0]
```

```python
import keras
from keras.layers import Input, Dense, SimpleRNN, Embedding
model = keras.Sequential()
model.add(Input(shape=(100,)))
model.add(Embedding(200, 16))
model.add(SimpleRNN(8))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
model.fit(X_train, y_train, epochs=100, batch_size=32, callbacks=[EarlyStopping(patience=3, restore_best_weights=True)])
```

```python
model.evaluate(X_test, y_test)
```
