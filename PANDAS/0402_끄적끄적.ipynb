```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.get_dataset_names()
```

```python
# Q1. dataset 에서 "passengers" 부터 "taxis" 를 불러오는데 payment 까지만 열고 아래처럼 그래프를 만들어 봅시다.

taxis = sns.load_dataset("taxis").loc[:, 'passengers' :'payment']
taxis
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

sns.violinplot(data=taxis, x='payment', y= 'total', hue='color', dodge=False, ax=ax[0])
sns.boxplot(data=taxis, x='payment', y= 'total', hue='color', dodge=False, ax=ax[1])
```

```python
# Q2. 하나를 더 추가해보자.

import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

sns.violinplot(data=taxis, x='payment', y='total', hue='color', dodge=False, ax=ax[0])
ax[0].set_title('Violin Plot')

sns.boxplot(data=taxis, x='payment', y='total', hue='color', dodge=False, ax=ax[1])
ax[1].set_title('Box Plot')

sns.stripplot(data=taxis, x='payment', y='total', hue='color', dodge=True, ax=ax[2], alpha=0.3)
ax[2].set_title('Strip Plot')

plt.tight_layout()
plt.show()
```
