```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Q1. sns.load_dataset 에서 planets 를 불러와 연도 별로 orbital_period 의 
# mean (평균) 
# median (중앙값)
# max (최댓값)
# count (개수) 를 출력해보자

planets = sns.load_dataset('planets')

stats = planets.groupby('year')['orbital_period'].agg(['mean', 'median', 'max', 'count']).reset_index()
stats
```

```python
# Q2. 그래프 그리기
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Mean
ax[0,0].plot(stats['year'], stats['mean'], 'g')
ax[0, 0].set_title('Average Orbital Period (Mean)')
ax[0, 0].grid(True, alpha=0.3)

# Median
ax[0, 1].fill_between(stats['year'], stats['median'], color='pink', alpha=0.5)
ax[0, 1].plot(stats['year'], stats['median'], color='skyblue', linewidth = 2)
ax[0, 1].set_title('Typical Orbital Period (Median Area)')

# Maximum
sns.regplot(data=planets, x='year', y='orbital_period', ax=ax[1,0], color='purple')
ax[1,0].set(yscale='log', title='Orbital Period Trend')

# Count
sns.barplot(data=stats, x='year', y='count', hue='year', palette='pastel', ax=ax[1, 1], legend=False)
ax[1, 1].set_title('Number of Planets Discovered')
ax[1, 1].tick_params(axis='x', labelrotation=45)

plt.tight_layout()
plt.show()
```

```python
# Q3. 가장 많이 발견된 해(Peak Year) 추출

peak_year = stats.loc[stats['count'].idxmax(), 'year']
print(f"가장 많은 행성이 발견된 해: {peak_year}년")
```
